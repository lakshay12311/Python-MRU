Q3. Calculate factorial of a number using function?

```python
n = int(input("Enter the number : "))

def factorial(n):
    total = 1
    for i in range(1,n+1):
        total *= i
    return total

print("The factorial of " + n + " is " + factorial(n))
```

Q5. Write a program to swapping of two numbers using functions?
