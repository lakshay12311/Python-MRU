##Q13. Concatenate three dictionaries
##Sample Dictionary :
##dic1={1:10, 2:20}
##dic2={3:30, 4:40}
##dic3={5:50,6:60}
##Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

```python
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
dic4 = {}
for d in (dic1, dic2, dic3): dic4.update(d)
print(dic4)
```


##Q15. Insert factorial of keys in values. And print dictionary?

```python
d = {}

def fact(i):
    factorial = 1
    for i in range(1,i):
        factorial *= i
    return factorial
        
for i in range(1,11):
    d[i] = fact(i)
    
print(d)
```
