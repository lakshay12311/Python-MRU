
Write a Python program to demonstrate Inheritance and method overriding.

```python
class Parent1():
        
    def show(self):
        print("Inside Parent1")
          
class Parent2():
          
    def display(self):
        print("Inside Parent2")
          
class Child(Parent1, Parent2):
        
    def show(self):
        print("Inside Child")
  
obj = Child()
obj.show()
obj.display()
```

Write a Python program to demonstrate multiple Inheritance.


```python
class Class1:
    def m(self):
        print("In Class1")
       
class Class2(Class1):
    def m(self):
        print("In Class2")
 
class Class3(Class1):
    def m(self):
        print("In Class3") 
        
class Class4(Class2, Class3):
    pass  
     
obj = Class4()
obj.m()
```
